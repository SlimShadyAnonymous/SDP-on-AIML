```python
import pandas as pd
import numpy as np
bigmart = pd.read_csv("bigmart.csv")
bigmart.head()
for n in bigmart.columns:
    if(bigmart[n].isnull().mean()*100 > 20):
        bigmart.drop(n, axis = 1,inplace = True)
for n in list(bigmart.index):
    if(bigmart.loc[n].isnull().mean()*100 > 20):
        bigmart.drop(n, axis = 0,inplace = True)
bigmart.head()
```

```python
# mean for item weight column
# mode for outlet_size column

bigmart["Item_Weight"].fillna(bigmart["Item_Weight"].mean(), inplace = True)
```

```python
placement = pd.read_csv("placement.csv")
placement
```

```python
sns.distplot(placement["cgpa"])
```

```python
sns.distplot(placement["placement_exam_marks"])
```

```python
print(placement["cgpa"].mean())
print(placement["cgpa"].std())
```

```python
placement["cgpa"].describe
```

```python
#upper & lower values
upper_value = 3 * placement["cgpa"].std() + placement["cgpa"].mean()
lower_value = -3 * placement["cgpa"].std() + placement["cgpa"].mean()
print (upper_value, lower_value)
```

```python
def zscore(x):
    if x < lower_value:
        return lower_value
    elif x > upper_value:
        return upper_value
    else:
        return x
placement["cgpa"] = placement["cgpa"].apply(zscore)
```

```python
sns.distplot(placement["cgpa"])
```

```python
placement["placement_exam_marks"].describe()
```

```python
sns.boxplot(placement["placement_exam_marks"])
```

```python
placement["placement_exam_marks"].quantile(0.5)
```

```python
Q1 = placement["placement_exam_marks"].quantile(0.25)
Q3 = placement["placement_exam_marks"].quantile(0.75)
IQR = Q3 - Q1
upper_value = Q3 + 1.5 * IQR
lower_value = Q1 - 1.5 * IQR
print(upper_value, lower_value)
```

```python
def iqr(x):
    if x < lower_value:
        return lower_value
    elif x > upper_value:
        return upper_value
    else:
        return x
placement["placement_exam_marks"] = placement["placement_exam_marks"].apply(iqr)
```
